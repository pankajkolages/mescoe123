```python
Data Analytics I
Create a Linear Regression Model using Python/R to predict home prices using Boston Housing
Dataset (https://www.kaggle.com/c/boston-housing). The Boston Housing dataset contains
information about various houses in Boston through different parameters. There are 506 samples
and 14 feature variables in this dataset.
The objective is to predict the value of prices of the house using the given features.
```

```python
import pandas as pd
```

```python
df= pd.read_csv('C:/Users/jyoti/Documents/Engineering Third Year/DSBDAL/datasets/housing.csv')
```

```python
df.head()
```

```python
missing_values= df.isnull().sum()
print(f"Missing values in dataset:\n{missing_values}")
```

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']]
Y = df['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state= 42)

print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {Y_train.shape}")
```

```python
from  sklearn.linear_model import LinearRegression

model= LinearRegression()
model.fit(X_train,Y_train)
```

```python
Y_pred= model.predict(X_test)
print(f"Predicted values for house prices:\n{Y_pred}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y_test, Y_pred)
print(f"Mean Squared Error:\n{mse}")
```

```python
r2 = r2_score(Y_test, Y_pred)
print(f"Root Squared:\n {r2}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(Y_test, Y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.show()
```
